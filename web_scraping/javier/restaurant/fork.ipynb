```python
def get_menu(driver):
    patter = re.compile(r'\d+(\,?\d+)? €')
    wait = WebDriverWait(driver,1200)
    menu = wait.until(EC.visibility_of_element_located((By.XPATH,'//dl[@class="css-1ndrudv e1xxesyf0"]/div[@class="css-1iqq28q elkhwc30"]//span')))
    menu = [i.text for i in menu.find_elements_by_xpath('//dl[@class="css-1ndrudv e1xxesyf0"]/div[@class="css-1iqq28q elkhwc30"]//span')]
    
    menu_text = []
    menu_price = []
    for i in menu:
        if re.match(patter,i):
            menu_price.append(i)
        else:
            menu_text.append(i.strip())
        
    #Eliminar repetidos
    box = []
    for i in menu_text:
        if i.strip() not in box:
            box.append(i.strip())
        else:
            pass
        
    #Concatenar
    menu = list(zip(box,menu_price))
    menu = [' -- '.join(i) for i in menu]
    return menu
```

```python
def get_info(r,header):
    soup = BeautifulSoup(r.text, 'lxml')
    title = soup.find('h1').get_text()

    try:
        direccion = soup.find('span',attrs={'class':"eeio5r41 css-1juo9mb e1xxesyf0"}).find('span',attrs={'class':"css-1lnlsqc eulusyj0"}).text
    except:
        direccion = None
    try:
        descuento = remove_blank_spaces(soup.find('div',attrs={'class':"css-3o5z22 e1xxesyf0"}).find('p').text.replace(',','|')).replace('|',',')
    except:
        descuento = None
    try:
        mean_price = remove_blank_spaces(soup.find_all('span',text='Precio medio')[0].find_next_sibling('span').text)
    except:
        mean_price = None
    try:
        image = soup.find('div',attrs={'class':"slick-track"}).find('img').get('src')
        image_name = download_images.download_image_with_requests(image,nombre_del_lugar='fork_j',title_for_image='_'.join(title.lower().strip().split()))
    except:
        image_name = None
    try:
        link = soup.find('div',attrs={'class':"css-ld4rsw elkhwc30"}).a.get('href')
        link_menu = 'https://www.thefork.es{}'.format(link)
        driver.get(link_menu)
        menu = get_menu(driver)

        assert (menu != []),"Error menu vacio AS"
        menu = '  /  '.join(menu)

        
    except NoSuchElementException as nx:
        print(nx)
        input('Press a key: ')
        menu = get_menu(driver)
        assert (menu == []),"Error menu vacio AS"
    
    except AssertionError as ax:
        print(ax)
        input('Press a key: ')
        menu = get_menu(driver)
        
        
    except Exception as ex:
        print('Error en menu: ',ex)
        traceback.print_exc()
        menu = None
        link_menu = None
    try:
        categoria = soup.find('p',attrs={'class':"css-zmb4ti eulusyj0"}).find_all('span',attrs={'class':"css-17qrp4o e1xxesyf0"})
        categoria = [a.text for i in categoria for a in i.find_all('span', attrs={'data-test':True}) if a['data-test'] == 'restaurant-page-restaurant-tags-CUISINE']
        categoria = categoria[0]
    except:
        categoria = 'None'
    return header , {
            'Nombre':title,
            'Direccion':direccion,
            'Precio Medio':mean_price,
            'Descuento':descuento,
            'Categoria':categoria,
            'Image':image_name,
            'Menu':menu,
            'link_menu':link_menu
        }
```

```python
data_links = []

for page in range(1,23):
    for i in lines:
        i = i.replace('\n','')
        proxy = i
        link = 'https://www.edarabia.com/schools/uae/?pg={}'.format(page)
        try:
            r = requests.get(link,proxies={'http':proxy,'https':proxy},timeout=3)
            soup = BeautifulSoup(r.text,'lxml')
            print(r.status_code)
            if soup.find_all('div',attrs={'class':"list-items"}):
                list_links = [i.find('div',attrs={'class':"row"}).find('div',attrs={'class':"col-md-10 col-sm-10 col-xs-10"}) for i in soup.find_all('div',attrs={'class':"list-items"}) if i is not None]
                links_page = [i.h5.a.get('href') for i in list_links if i is not None]
                data_links.append(links_page)
                break
                
        except:
            print('Failed')
```

```python
link_eua = list(set([a for i in data_links for a in i]))
```
