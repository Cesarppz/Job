We bring another very important column for our analysis, which is an estimate of how much is the annual profit of developers.

This column is a range. To make an interpolation of the missing data, we have to convert the column to a numeric column. Being a range we will take the average between the two ranges. 

```python
def transform_str_int(x):
    if x != np.nan:
        if type(x) == float:
            pass
        else:
            if x == '> $500,000':
                ramdon_num = 500.000
                return ramdon_num
            else:
                inf, sup = x.split('-')
                inf, sup = float(inf.replace(',','').replace('$','').strip()), float(sup.replace(',','').replace('$','').strip())
                random_num = round(abs(np.mean([inf,sup])),2)
#                 ramdon_num = round(abs(stats.norm.rvs(inf,sup)),2)
                return random_num
```

```python
numeric_df['year_compesation'] = numeric_df['year_compesation'].apply(lambda x : transform_str_int(x))
```

```python
numeric_df['year_compesation'].isna().sum()
```

```python
numeric_df['year_compesation'] = numeric_df['year_compesation'].interpolate()
```

```python
numeric_df['year_compesation'].isna().sum()
```

```python
numeric_df['category'] =pd.cut(numeric_df['year_compesation'],bins=10)
```

```python
numeric_df['category'] = pd.Categorical(numeric_df.category)
```

Once the column is completely clean we can make the analysis. But first we put the figures together in range for better handling.

In the data frame to differentiate by gender, it is necessary to clean up and unify the different terms used.

```python
numeric_df.drop(columns=['What is your gender? - Prefer to self-describe - Text'],inplace=True)
```

```python
def change_gender(x):
    if x == 'Male':
        return 'Man'
    elif x == 'Female':
        return 'Woman'
    else:
        return x
```

```python
numeric_df['gender_u'] = numeric_df.loc[:,'gender'].apply(lambda x: change_gender(x))
```

## Country and gender

```python
numeric_df['country'] = df['In which country do you currently reside?']
```

```python
num_person_per_country_man = numeric_df.query('gender_u == "Man"')['country'].value_counts().values.sum()
num_person_per_country_woman = numeric_df.query('gender_u == "Woman"')['country'].value_counts().values.sum()
```

```python
fig, (ax1,ax2) = plt.subplots(2,1,figsize=(28,13),sharex=True,sharey=True)
#plt.yticks(ticks = range(0,30,5) , labels=[str(i) + ' %' for i in range(0,30,5) ])

plt.setp( (ax1,ax2), yticklabels=[str(i) + ' %' for i in range(0,35,5)],yticks=range(0,35,5))
                                 
ax1.bar(numeric_df.query('gender_u == "Man"')['country'].value_counts().index,[i / num_person_per_country_man *100 for i in numeric_df.query('gender_u == "Man"')['country'].value_counts().values], alpha=0.6)
ax1.tick_params(rotation=90,axis='x')
ax1.set_title('Countries where men reside')
ax1.set_xlabel('Country')
ax1.set_ylabel('Number of men')

ax2.bar(numeric_df.query('gender_u == "Woman"')['country'].value_counts().index,[i / num_person_per_country_woman *100 for i in numeric_df.query('gender_u == "Woman"')['country'].value_counts().values], alpha=0.9,color='red')
ax2.tick_params(rotation=90,axis='x')
ax2.set_title('Countries where women reside ')
ax2.set_xlabel('Country')
ax2.set_ylabel('Number of women')

fig.suptitle('Analysis of countries by gender',fontsize=32)

plt.show()
```

The most remarkable thing about this graph is that we can see that in percentage terms, the number of men and women does not vary much, with exceptions such as the USA where we have a 5% difference between male and female participants.
